## Create a dataframe of movie titles, including Bond movies

```python
# import dependencies
import pandas as pd
```

```python
# get movie titles from Rotten Tomatoes for 
# action/ adventure movies and
# mystery / suspense movies

# scrape for table
url = 'https://www.rottentomatoes.com/top/bestofrt/top_100_action__adventure_movies/'
tables = pd.read_html(url)
action_table=tables[2]

url = 'https://www.rottentomatoes.com/top/bestofrt/top_100_mystery__suspense_movies/'
tables = pd.read_html(url)
suspense_table=tables[2]
```

```python
suspense_table
```

```python
suspense_table = suspense_table.drop(['Rank', 'RatingTomatometer', 'No. of Reviews'], axis = 1)
```

```python
suspense_table['title'] = suspense_table.Title.str.split("(", n=1, expand = True)[0]
suspense_table = suspense_table.drop(['Title'], axis = 1)
```

```python
# add column for bond movie, 1 for yes, 0 for no
# note there will be bond movies in this list, but we will remove them later
suspense_table['bond'] = 0
```

```python
# repeat for action table
action_table = action_table.drop(['Rank', 'RatingTomatometer', 'No. of Reviews'], axis = 1)
action_table['title'] = action_table.Title.str.split("(", n=1, expand = True)[0]
action_table['bond'] = 0
action_table = action_table.drop(['Title'], axis = 1)
```

```python
action_table
```

```python
# get list of Bond movies from Rotten Tomatoes by scraping page for xpath

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
driver = webdriver.Chrome(**executable_path)
```

```python
driver.get('https://www.rottentomatoes.com/franchise/james_bond_007')
time.sleep(2)
# used xpath to locate the elements
try:
    titles = driver.find_elements_by_xpath('//*[@id="franchise-media"]/div[2]/ul/li/div[2]/h3/a')
    
# exception for the times that the webpage does not load properly
except:
    print('something went wrong')


print(titles)
```

```python
# close the driver browser
driver.close()
```

```python
bond_title = []

for title in titles:
    bond_title.append(title.text)
```

```python
bond_title
```

```python
bond_table = pd.DataFrame(bond_title)
```

```python
bond_table.columns = ['title']
```

```python
# add column for bond movie, 1 for yes, 0 for no
bond_table['bond'] = 1
```

```python
bond_table
```

```python
# join tables together
title_table = pd.concat([action_table, suspense_table, bond_table])
```

```python
title_table
```

```python
title_table = title_table.loc[~title_table.title.duplicated(keep='last'), :]
# note there were two Bond movies called Casino Royale
```

```python
title_table
```

```python
# export to csv
title_table.to_csv('../Resources/title_table.csv',index = False)
```
